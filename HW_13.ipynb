### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv(r'C:\Users\Пользователь\Desktop\pyton dom\python_13_join\ml-latest-small\ratings_13.csv')
ratings.head()
```

```python

ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings['num'] = 1 
ratings = ratings.groupby('userId').agg({'date': ['min', 'max'], 'num': 'sum'})
ratings['aver_time'] = ratings['date']['max'] - ratings['date']['min']

ratings = ratings[ratings['num']['sum'] > 100]
ratings['aver_time'].mean()
```
